```python
import pandas as pd
df = pd.read_csv(r'C:\Users\LAVANYA\Downloads\archive (1).zip')
print(df)
```

```python
print("\nMissing Values:")#finding missing values
print(df.isnull().sum())
```

```python
df_filled = df.fillna(0)
print("\nDataFrame with filled missing values:")
print(df_filled)
```

```python
print("\nBasic Statistics:")
print(df.describe())
```

```python
sorted_df = df.sort_values('math score')#sorting
print("\nSorted DataFrame:")
print(sorted_df)
```

```python
filtered_df1 = df[df['math score'] <70]#filtering
filtered_df2 = filtered_df1[filtered_df1['reading score'] <70]#filtering
filtered_df3 = filtered_df2[filtered_df2['writing score'] <70]#filtering
print("\nFiltered DataFrame:")
print(filtered_df3)
```

```python
print("\nBasic Statistics:")
print(filtered_df3.describe())
```

```python
grouped_df =filtered_df3.groupby('writing score').mean()
print("\nGrouped DataFrame:")
print(grouped_df)
```
